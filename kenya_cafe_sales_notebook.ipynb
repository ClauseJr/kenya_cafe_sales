```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
sales = pd.read_csv("Sales_Transactions.csv")
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
staff = pd.read_csv("Staffs.csv")
```

```python
# Data Cleaning and Preprocessing

sales.info()
sales.shape
sales.head(5)
```

```python
sales.isnull().sum()
sales.duplicated().sum()
sales.drop_duplicates(inplace=True)
```

```python
sales.isnull().sum()
sales.shape

print(sales.head().to_string(max_rows=10))
```

```python
string_columns = ['CafeName', 'Branch', 'ProductName', 'Category','PaymentMethod']

for col in string_columns:
    sales[col] = sales[col].astype(str).str.strip().str.title()


id_columns = ['TransactionID', 'CustomerID', 'ProductID', 'StaffID']
for cols in id_columns:
    sales[cols] = sales[cols].astype(str).str.strip()

print(sales.to_string(max_rows=10))
```

```python
sales['CafeName'].value_counts()
sales['Branch'].value_counts()
sales['ProductName'].value_counts()
sales['Category'].value_counts()
sales['PaymentMethod'].value_counts()


cafe_map = {
    'CafeName':{
        "Art Caffe": "Art Caffe",
        "Java House": "Java House",
        "Artcafe": "Art Caffe",
        "Art Café": "Art Caffe",
        "Artcaffe": "Art Caffe"

    },

    'Branch' : {
        "Cbd" : "CBD",
        "Westside" : "Westlands",
        "Milimani" : "Kilimani"  
    },

    'PaymentMethod' : {
        "Mpesa" : "Mpesa",
        "M-Pesa" : "Mpesa"
    }
}

sales.replace(cafe_map, inplace=True)
print(sales.to_string(max_rows=10))
```

```python
sales['SaleDate'] = pd.to_datetime(sales['SaleDate'], dayfirst=True, errors='coerce')
print(sales['SaleDate'].head(10))
print(sales['SaleDate'].isna().sum())
```

```python
bad_dates = sales[sales['SaleDate'].isna()][['SaleDate']]
print(bad_dates.value_counts())

before = sales.shape[0]

sales = sales.dropna(subset=['SaleDate'])

after = sales.shape[0]
print(f"Rows dropped due to unparseable dates: {before - after}")
print(f"Rows remaining: {after}")


sales['Year'] = sales['SaleDate'].dt.year
sales['Month'] = sales['SaleDate'].dt.month
sales['MonthName'] = sales['SaleDate'].dt.strftime('%b')
sales['Quarter'] = sales['SaleDate'].dt.quarter.apply(lambda x: f"Q{x}")
sales['DayOfWeek'] = sales['SaleDate'].dt.strftime('%a')

print(sales[['SaleDate', 'Year', 'Month', 'MonthName', 'Quarter', 'DayOfWeek']].head(5))
```

```python
print(sales['SaleTime'].isna().sum())

sales['SaleTime'] = sales['SaleTime'].fillna('00:00')
sales['SaleTime'] = pd.to_datetime(sales['SaleTime'], format='%H:%M', errors='coerce')

sales['Hour'] = sales['SaleTime'].dt.hour

sales['TimeKnown'] = sales['Hour'] != 0

time_analysis = sales[sales['TimeKnown'] == True]

def time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

sales['TimeOfDay'] = sales['Hour'].apply(time_of_day)

sales['SaleTime'] = pd.to_datetime(sales['SaleTime']).dt.strftime('%H:%M')

print(sales.to_string(max_rows=10))
```

```python
sales['Discount_Pct'] = sales['Discount_Pct'].astype(str).str.replace('%','',regex=False).str.strip()
sales['Discount_Pct'] = pd.to_numeric(sales['Discount_Pct'], errors='coerce')

sales['Quantity'] = sales['Quantity'].astype(str).str.replace('-','',regex=False).str.strip()
sales['Quantity'] = pd.to_numeric(sales['Quantity'], errors='coerce')
sales['Quantity'] = sales['Quantity'].astype(int)

sales['UnitPrice_KES'] = pd.to_numeric(sales['UnitPrice_KES'], errors='coerce')

sales['Rating'] = pd.to_numeric(sales['Rating'], errors='coerce')
sales['Rating'] = sales['Rating'].fillna(sales['Rating'].median()).astype(int)


num_columns = ['Quantity', 'UnitPrice_KES', 'TotalAmount_KES','Discount_Pct', 'Rating']

print(sales[num_columns].dtypes)
```

```python
sales['TotalAmount_KES'] = sales['TotalAmount_KES'].fillna(
    sales['Quantity'] * sales['UnitPrice_KES'] * (1 - sales['Discount_Pct'] / 100)
)

print(sales['TotalAmount_KES'].isna().sum())
```

```python
sales.isnull().sum()

print("Null CustomerIDs:", sales['CustomerID'].isna().sum())

pct = (sales['CustomerID'].isna().sum() / len(sales)) * 100
print(f"Percentage of anonymous sales: {pct:.2f}%")

anon = sales[sales['CustomerID'].isna()]
print(anon[['TransactionID', 'CafeName', 'Branch', 'TotalAmount_KES', 'PaymentMethod']].head(10))
```

```python
sales['CustomerID'] = sales['CustomerID'].fillna('ANON')

print(sales['CustomerID'].isna().sum())
print(sales['CustomerID'].value_counts())
```

```python
print(sales.isnull().sum())
print(sales.duplicated().sum())

sales['SaleDate'] = sales['SaleDate'].dt.normalize()

sales.shape
sales.info()

sales['CustomerID'] = sales['CustomerID'].replace('ANON',np.nan)
```

```python
print(sales.to_string(max_rows=10))
```

```python
customers.info()
customers.shape

str_columns = ['CustomerName', 'Town', 'LoyaltyTier']

for cust_cols in str_columns:
    customers[cust_cols] = customers[cust_cols].astype(str).str.strip().str.title()

print(customers.to_string(max_rows=10))
```

```python
customers['JoinDate'] = pd.to_datetime(customers['JoinDate'], dayfirst=True, errors='coerce')

bad_dates = customers[customers['JoinDate'].isna()][['JoinDate']]
print(bad_dates.value_counts())

before = customers.shape[0]

customers = customers.dropna(subset=['JoinDate'])

after = customers.shape[0]
print(f"Rows dropped due to unparseable dates: {before - after}")
print(f"Rows remaining: {after}")
```

```python
customers.drop(columns=['Phone'], inplace=True)

print(customers.columns.tolist)
print(customers.to_string(max_rows=10))
```

```python
products.shape
products.info()
products.describe()

prod_columns = ['Category', 'ProductName']

for prod_col in prod_columns:
    products[prod_col] = products[prod_col].astype(str).str.strip().str.title()

print(products.to_string(max_rows=10))

products['UnitPrice_KES'] = pd.to_numeric(products['UnitPrice_KES'], errors='coerce')
```

```python
staff.info()
staff.shape

staff_columns = ['StaffName', 'Role', 'Branch']

for staff_col in staff_columns:
    staff[staff_col] = staff[staff_col].astype(str).str.strip().str.title()

print(staff.to_string(max_rows=10))

staff['Salary_KES'] = pd.to_numeric(staff['Salary_KES'], errors='coerce')
```

```python
staff['HireDate'] = pd.to_datetime(staff['HireDate'], dayfirst=True, errors='coerce')

bad_dates = staff[staff['HireDate'].isna()][['HireDate']]
print(bad_dates.value_counts())

before = staff.shape[0]

staff = staff.dropna(subset=['HireDate'])

after = staff.shape[0]
print(f"Rows dropped due to unparseable dates: {before - after}")
print(f"Rows remaining: {after}")
```
